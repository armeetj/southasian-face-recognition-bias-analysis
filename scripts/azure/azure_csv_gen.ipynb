```python
import csv
import pandas
import os
from tqdm import tqdm
import json
import time
```

```python
def run(directory):
    df = pandas.DataFrame(
        columns=["id", "predicted_gender"])
    counter = 0;
    for filepath in tqdm(os.listdir(directory + "json")):
        output_file = directory + "out.csv"
        in_filepath = directory + "json/" + filepath


        json_file = open(in_filepath, "r")
        try:
            json_file = json.loads(json_file.read())
            gender  = json_file[0]["faceAttributes"]["gender"]
            print(gender)
            df = df.append([
                {
                    "id": counter,
                    "predicted_gender": gender,
                }])
        except Exception:
            print(in_filepath)
        counter += 1
    df.to_csv(output_file)
```

```python
run("../../data2/2_north_ladakh/out/azure/male/")
run("../../data2/2_north_ladakh/out/azure/female/")

run("../../data2/3_north_punjab/out/azure/male/")
run("../../data2/3_north_punjab/out/azure/female/")

run("../../data2/4_northwest_rajasthan/out/azure/male/")
run("../../data2/4_northwest_rajasthan/out/azure/female/")

run("../../data2/5_east_jharkhand/out/azure/male/")
run("../../data2/5_east_jharkhand/out/azure/female/")

run("../../data2/6_south_telangana/out/azure/male/")
run("../../data2/6_south_telangana/out/azure/female/")

run("../../data2/7_deepsouth_tamil_nadu/out/azure/male/")
run("../../data2/7_deepsouth_tamil_nadu/out/azure/female/")

run("../../data2/8_west_gujrat/out/azure/male/")
run("../../data2/8_west_gujrat/out/azure/female/")
```
